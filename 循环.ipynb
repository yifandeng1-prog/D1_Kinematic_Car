```python
#FOR循环
for i in range(5):
    print(i)
for r in range(1,6):
    print(r) 
```

```python
list=["a","b","c"]
for item in list:
        print(item)
```

```python
#WHILE循环
#注意count =count+1，不然程序会卡死
count=0
while count <= 5:
    print(count)
    count =count+1
```

```python
#1-10奇数
for i in range(1,11):
    if i %2==0:
        print(i)
```

```python
#1-100的和
#记得初始化变量
total=0
count=1
while count <101:
    total=total+count
    count=count+1
print(total)
```

```python
numbers=[3, 7, 2, 9, 5, 1, 8]
for num in numbers:
    if num == 5:
        break
    if num % 2 == 0:
        continue
    print(num)
```
