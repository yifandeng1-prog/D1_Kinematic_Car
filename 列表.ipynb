```python
'''
有一个列表fruits = ["苹果", "香蕉", "橙子", "葡萄", "西瓜"]，请：
打印出第二个元素
打印出最后一个元素
打印出前三个元素
'''
fruits = ["苹果", "香蕉", "橙子", "葡萄", "西瓜"]
print(fruits[1])
print(fruits[-1])
print(fruits[0:3])
```

```python
'''
有一个空列表numbers = []，请：
向列表中添加数字 5、3、8、1、9
删除列表中的第三个元素
对列表进行升序排序
打印最终的列表
'''
numbers=[]
numbers.append(5)
numbers.append(3)
numbers.append(8)
numbers.append(1)
numbers.append(9)
print(numbers)
numbers.pop(2)
print(numbers)
numbers.sort()
print(numbers)
```
