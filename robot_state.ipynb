```python
class RobotState:
    def __init__(self,x=0.0,y=0.0,yaw=0.0,v=0.0):
        self.x=x
        self.y=y
        self.yaw=yaw
        self.v=v

def print_state(state):
    print(f"x = {state.x:.2f}, y = {state.y:.2f}, yaw = {state.yaw:.2f}, v = {state.v:.2f}")


if __name__ == "__main__":
    robot = RobotState(x=1.0, y=2.0, yaw=0.5, v=1.2)

    print_state(robot)

        
```
