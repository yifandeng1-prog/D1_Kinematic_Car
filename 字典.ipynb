```python
'''
有一个学生信息字典student = {"name": "张三", "age": 18, "score": 95}，请：
打印学生的姓名
用get()方法打印学生的性别（如果不存在，返回 "未知"）
'''
student = {"name": "张三", "age": 18, "score": 95}
print(student["name"])
print(student.get("gender", "unknown"))
```

```python
'''
有一个空字典person = {}，请：
添加键值对："name": "李四"，"age": 20，"city": "北京"
修改年龄为 21
遍历字典，打印所有键和值
'''

person = {}
person["name"]="李四"
person["age"]=20
person["city"]="北京"
print(person)
person["age"]=21
print(person)
for key,value in person.items():
    print(f"{key},{value}")
```
