```python
'''
定义一个名为Dog的类，包含两个属性：名字（name）和品种（breed）。然后创建两个 Dog 对象：一只叫 "旺财" 的金毛，一只叫 "小白" 的泰迪，并打印它们的名字和品种。
'''

class dog:
    def __init__(self,name,bread):
        self.name=name
        self.bread=bread
dog1=dog("旺财","金毛")
dog2=dog("小白","泰迪")

print(dog1.name,dog1.bread)
print(dog2.name,dog2.bread)
```

```python
'''
给上面的Dog类添加一个名为bark的方法，功能是打印 "XX（名字）在汪汪叫！"。然后创建一个 Dog 对象，调用它的bark方法。
'''

class dog:
    def __init__(self,name,bread):
        self.name=name
        self.bread=bread

    def bark(self):
        print(f"{self.name}在汪汪叫！")

dog1=dog("旺财","金毛")
dog2=dog("小白","泰迪")

print(dog1.name,dog1.bread)
print(dog2.name,dog2.bread)

my_dog = dog("大黄", "中华田园犬")
my_dog.bark()
```
