```python
'''
*是指每个元素分别相乘A∗B=[a11​b11   ​a21​b21     @是指矩阵乘法
                        ​​a12​b12   ​a22​b22​​]
'''


import numpy as np

# 定义二维点
p = np.array([1.0, 0.0])  # 点 (1,0)

# 定义旋转角度（弧度）
theta = np.pi / 4  # 45度

# 定义二维旋转矩阵 R
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

# 旋转点
p_rot = R @ p

# 输出结果
print("原始点:", p)
print("旋转后点:", p_rot)
```
