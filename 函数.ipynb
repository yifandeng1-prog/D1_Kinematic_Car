```python
'''
定义一个名为print_separator的函数，功能是打印一行由 20 个-组成的分隔线。然后调用这个函数 3 次。
'''

def print_separator():
    print("-"*20)
print_separator()
print_separator()
print_separator()
```

```python
'''
定义一个名为calculate_rectangle_area的函数，接收两个参数：长和宽，返回长方形的面积。然后调用这个函数，计算长为 6、宽为 4 的长方形面积，并打印出来
'''

def calculate_rectangle_area(length,wieth):
    area=length*wieth
    return area
result=calculate_rectangle_area(6,4)
print(result)
```
